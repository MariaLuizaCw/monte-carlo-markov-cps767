```python
import random
import matplotlib.pyplot as plt
import math

# Algoritmo do item 1
def gerar_V():
    cont = 0
    k = 0
    
    while cont < 1:
        U = random.uniform(0, 1)
        cont += U
        k += 1
    
    return k


# Estimativa de Monte Carlo para um n fixo
def estimar_E_V(n):
    soma = 0
    
    for i in range(n):
        soma += gerar_V()
    
    return soma / n


# Gera a convergência variando n
def gerar_convergencia(max_n):
    estimativas = []
    
    for n in range(1, max_n + 1):
        estimativa = estimar_E_V(n)
        estimativas.append(estimativa)
    
    return estimativas


# Parâmetro
max_n = 2000  # pode aumentar depois

estimativas = gerar_convergencia(max_n)

# Plot
plt.figure()
plt.plot(range(1, max_n + 1), estimativas, label='Estimativa Monte Carlo')
plt.axhline(y=math.e, linestyle='--', label='Valor esperado (e ≈ 2.718)')
plt.xlabel('Número de amostras (n)')
plt.ylabel('Estimativa de E[V]')
plt.title('Convergência da estimativa de E[V]')
plt.legend()
plt.grid()

plt.show()
```
